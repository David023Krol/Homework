## Způsob 1: Použití knihovny numpy

Používáme zabudovanou knihovnu random a konstrukci zvanou list comprehension, která vytvoří dvourozměrný seznam. Každý prvek je náhodné celé číslo z intervalu 0 až 10.

```python
import numpy as np

n = 5

matrix = np.random.randint(0, 11, size=(n, n))
print(matrix)
```

## Způsob 2: Použití knihovny random a list comprehension

NumPy je výkonná knihovna pro práci s poli. Pomocí funkce np.random.randint() vygenerujeme matici s náhodnými čísly v daném rozsahu a rozměru. Tento způsob je rychlý a efektivní, zejména pro větší matice.

```python
import random

n = 5  

matrix = [[random.randint(0, 10) for _ in range(n)] for _ in range(n)]
for row in matrix:
    print(row)
```

## Způsob 3: Pomocí pandas DataFrame

Pandas je knihovna pro práci s tabulkovými daty. Pomocí DataFrame můžeme uložit matici v přehledné tabulkové podobě. Náhodná čísla generujeme přes NumPy, ale výstup má formu dobře čitelné tabulky.



```python
import pandas as pd

n = 5 

df = pd.DataFrame(np.random.randint(0, 11, size=(n, n)))
print(df)
```
